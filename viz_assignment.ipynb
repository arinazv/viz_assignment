```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv(r'C:\Users\lenov\OneDrive\Рабочий стол\eurovision.csv')
```

```python
data.columns.tolist()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.scatter(x = data['running_order'], y = data['rank'])
plt.xlabel("Order")
plt.ylabel("Rank of the artist")

fig.set_figwidth(18) 

plt.show()

fig = ax.get_figure()
fig.savefig("figscatter.pdf")
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
newdata = data[['year', 'artist_country', 'rank' ]]

newdata.head()
```

```python
res = pd.pivot_table(newdata, index = ['artist_country'],columns =['year'], values = ['rank'])
ax = sns.heatmap(res)
ax.set_xlabel('Year', font='serif', fontsize=12)
ax.set_ylabel('Country', font='serif', fontsize=12)


fig = ax.get_figure()
fig.savefig("figheatmap.pdf")
```
